### RFM Modelling
Engineer below features for each individual user
1. Recency   : No of days since the last purchase date (today_date - last_invoice_date)
2. Frequency : Total number of transactions made by the customer
3. Monetary  : Total value of transactions made by the customer

```python
# making a copy of the dataframe
temp = df.copy()
temp['CustomerID'] = temp['CustomerID'].astype("object")
```

##### Recency

```python
# computing the last purchase feature of the users
temp['LastPurchase'] = temp.groupby("CustomerID")["InvoiceDate"].transform(lambda x: x.max(skipna=True))

# getting the last invoice
last_invoice = temp['InvoiceDate'].max() + pd.Timedelta(days=1)

# compuint the recency feature
temp['Recency'] = (last_invoice-temp['LastPurchase']).dt.days

# creating a separate dataframe which consists receny for each users
users_receny = temp[["CustomerID","Recency"]].drop_duplicates().sort_values(by="CustomerID").set_index("CustomerID")
users_receny.head()
```

##### Frequency

```python
users_frequency = pd.DataFrame(temp['CustomerID'].value_counts().sort_index()).rename(columns={"count":"Frequency"})
users_frequency.head(5)
```

##### Monetary

```python
users_monetary = pd.DataFrame(temp.groupby("CustomerID")['TotalSales'].sum().sort_index()).rename(columns={"TotalSales": "Monetary"})
users_monetary.head()
```

##### RFM Dataframe 

```python
# Joining Recency and Frequency DataFrames
RFM = pd.merge(users_receny, users_frequency, left_index=True, right_index=True)
# Joining with Monetary DataFrame
RFM = pd.merge(RFM, users_monetary, left_index=True, right_index=True).reset_index()

RFM.head()
```

### Creating RFM Ranks

To Conduct RFM Analysis, we need to rank our customers/users based on each RFM attribute separately, In this project let us assume we rank users from 1-4 using RFM values

1. We need to sort the Recency dataframe by most recent purchases at the top (in ascending order) and assign R score from 1-4 ( more recent more R score )

```python
# definnina a function which assings the score based on the threshold
def assign_score(x,quants,attribute):
    percentile25 = quants[0.25]
    percentile50 = quants[0.5]
    percentile75 = quants[0.75]

    # low recent ==> value high R score
    if attribute == "Recency":
        if x<=percentile25:
            return 4
        elif (x>percentile25) & (x<=percentile50):
            return 3
        elif (x>percentile50) & (x<=percentile75):
            return 2
        else:
            return 1

    # low frequency/monetary ==> low F/M score
    elif attribute in ("Frequency","Monetary"):
         if x<=percentile25:
            return 1
         elif (x>percentile25) & (x<=percentile50):
            return 2
         elif (x>percentile50) & (x<=percentile75):
            return 3
         else:
            return 4
        


# we need to assing a score from 1-4 so we need 3 quantiles for each attributes of RFM
recency_quantiles   = RFM['Recency'].quantile([.25,.50,.75]).to_dict()
frequency_quantiles = RFM['Frequency'].quantile([.25,.50,.75]).to_dict()
monetary_quantiles  = RFM['Monetary'].quantile([.25,.50,.75]).to_dict()
```

```python
# assinging R score for recency
RFM['R'] = RFM['Recency'].apply(lambda x:assign_score(x, recency_quantiles, "Recency"))

# assinging F score for frequency
RFM['F'] = RFM['Frequency'].apply(lambda x:assign_score(x, frequency_quantiles, "Frequency"))

# assinging M score for Monetary
RFM['M'] = RFM['Monetary'].apply(lambda x:assign_score(x, monetary_quantiles, "Monetary"))
```

```python
RFM.head()
```

We have 4 different scores for each attribute and we have 3 attributes so possible combinations are 4**3=64
For example if a customer has R,F,M scores of 1,1,1 then he will fall in 111 category like this each user may fall under 64 possible combinations

```python
# finding the in which group each user falls
RFM['Group'] = RFM['R'].apply(str) + RFM['F'].apply(str) + RFM['M'].apply(str)
RFM.head()
```

We have 64 possible groups, now we need to create 4 groups Bronze, Silver, Platinum, Gold  by combine this 64 groups 

### RFM Score

Now we need to aggregate R,F and M scores and create RFM score

```python
# creating RFM scores
RFM['Score'] = RFM[['R','F','M']].sum(axis=1)
RFM.head(5)
```

Now based on the Score we need to Categorize our customers into 4 Categories

```python
# defining our 4 groups interms of loyalty
loyalty = ["Bronze", "Silver", "Gold", "Platinum"]
# creating a column called Loyalty
RFM['Loyalty'] = pd.qcut(RFM['Score'], q=4, labels=loyalty)
```
